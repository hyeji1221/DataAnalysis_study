**중요한 몇 개 기사만 텍스트 파일로 저장하기**

```python
import os, re
import urllib.request as ur
from bs4 import BeautifulSoup as bs
```

```python
news = 'https://news.daum.net/'
soup = bs(ur.urlopen(news).read(), 'html.parser')
```

## 머리기사 제목 추출하기
- 하나라도 정확하게 작동하도록 만들고 그다음에 반복하는 코드 만들기
1. div 태그 속성값 확인하기
2. find_all로 div 내용 추출하기
3. 반복문으로 기사 제목 모두 추출하기 

```python
soup.find_all('div',{"class" : "item_issue"})
```

```python
for i in soup.find_all('div',{"class" : "item_issue"}):
    print(i.text)
```

## 하이퍼링크 주소 추출하기
- a 태그는 하이퍼링크를 만들 때 사용하는 태그
- `<a href = "링크할 URL 주소">하이퍼링크 텍스트</a>`

```python
soup.find_all('a') # <a> 태그만 추출하기
```

```python
for i in soup.find_all('a')[:5]:
    print(i.get('href')) # href 속성값만 가져오기
```

## 원하는 영역에서  하이퍼링크 모두 추출하기

```python
news = 'https://news.daum.net/'
soup = bs(ur.urlopen(news).read(), 'html.parser')
```

```python
for i in soup.find_all('div', {"class" : "item_issue"}):
    print(i.find_all('a')) # <a> 태그 추출하기
```

```python
for i in soup.find_all('div', {"class":"item_issue"}):
    print(i.find_all('a')[0].get('href'))
```

### 하이퍼링크 추출 과정
1. soup 객체에서 class 속성값이 item_issue인 div 태그를 find_all로 가져온다
2. div 태그 안에서 a 태그를 find_all로 가져온다
3. a 태그의 href 속성값을 get으로 출력한다
- find_all : 태그를 출력
- get : 태그 안에 있는 속성의 값을 추출

## 기사 제목과 내용 한꺼번에 추출하기

### 기사 내용 가져오기

```python
article1 = 'https://news.v.daum.net/v/20210115053146781'
soup2 = bs(ur.urlopen(article1).read(), 'html.parser')
```

```python
for i in soup2.find_all('p'): 
    print(i.text)
```

### 기사 제목 가져오기

```python
headline = soup.find_all('div', {"class": "item_issue"})
print(headline[0].text)
```

### 하이퍼링크된 모든 기사의 제목과 본문 추출하기

```python
for i in headline:
    print(i.text, '\n')
```

### 제목을 출력할 때 마다 그 기사의 내용 함께 출력하기

```python
for i in headline:
    print(i.text,'\n') # 기사 제목을 출력하기
    soup3 = bs(ur.urlopen(i.find_all('a')[0].get('href')).read(), 'html.parser') # 해당 기사가 올라와 있는 웹 사이트를 열어 soup3 객체에 저장
    for j in soup3.find_all('p'):
        print(j.text)
```
